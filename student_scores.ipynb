```python
import matplotlib.pyplot as plt

gender_count = df['gender'].value_counts()

itemName = gender_count.index
itemValue = gender_count.values
colors = ['skyblue', 'pink']
explode = (0, 0.1)

plt.figure(figsize=(6, 6)) 

plt.pie(itemValue, labels=itemName, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle = 140)

plt.title('The counts of genders', fontsize='x-large')

plt.show()
```

```python
import matplotlib.pyplot as plt

activities_count = df['extracurricular_activities'].value_counts()

itemName = activities_count.index
itemValue = activities_count.values
colors = ['salmon', 'lightgray']
explode = (0, 0)

plt.figure(figsize=(6, 6)) 

plt.pie(itemValue, labels=itemName, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle = 140)

# donut pie chart
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.title('The counts of activities by students', fontsize='x-large')

plt.show()
```

```python
career_aspirat_count = df['career_aspiration'].value_counts()

A = career_aspirat_count.index
B = career_aspirat_count.values
colors = ['#1f77b4', '#396ab1', '#5169ac', '#6968a7', '#8167a3', '#99669e', '#b16699', '#c86594', '#e0638f', '#87ceeb',
                '#a7d6e6', '#c6dfea', '#e6f0ff', '#aac0d1', '#89a7bf', '#6c8eab', '#4e7598']

plt.figure(figsize=(10,6))
plt.barh(A, B, color = colors)

plt.title('Career Inspiration for students', fontsize = 'xx-large')
plt.xlabel('counts')
plt.ylabel('carrer Inspiration')
plt.xticks(rotation = 45)

for i in range(len(A)):
    plt.text( B[i]+ 0.5, i, B[i], ha='left', va='center')

    plt.gca().invert_yaxis()  

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing student scores
# Calculate the average of each column

average_scores = df.mean()

itemName = average_scores.index
itemValue = average_scores.values
colors = ['#4169E1', '#3CB371', '#FFD700', '#9370DB', '#FF69B4', '#FFA500', '#808080']


# Plotting
plt.figure(figsize=(10, 6))

plt.bar(itemName,itemValue, color= colors)

plt.title('Average Scores of Subjects')
plt.xlabel('Subjects')
plt.ylabel('Average Score')


plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
